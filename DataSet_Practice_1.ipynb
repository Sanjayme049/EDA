```python
def discount_range(i):
        if i in range(0,26):
            return str("0-25%")
        elif i in range(26,51):
            return str("26-50%")
        elif i in range(51,76):
            return str("51-75%")
        else:
            return str(">75%")
df["Discount_range"] = df["Discount"].apply(discount_range)
df.head()
```

```python
df.rename(columns = {"Discount":"Discount_in_%"}, inplace = True)
```
